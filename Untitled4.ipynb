```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r"C:\Users\Mritunjoy Paul\CSV files\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
# Clean TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(subset=['TotalCharges'], inplace=True)
```

```python
df.head(2)
```

```python
# Drop ID, encode target
df.drop("customerID", axis=1, inplace=True)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
# One-hot encode categorical columns
df = pd.get_dummies(df, drop_first=True)
```

```python
# Split features/target
X = df.drop('Churn', axis=1)
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

```python
for name, model in models.items():
    print(f"ðŸ”· Model: {name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("ROC-AUC:", roc_auc_score(y_test, y_proba))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\n" + "-"*50 + "\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use best model (example: Random Forest)
importances = models['Random Forest'].feature_importances_
feat_importance = pd.Series(importances, index=X.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=feat_importance[:10], y=feat_importance.index[:10])
plt.title("Top 10 Feature Importances (Random Forest)")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```

```python
from matplotlib.backends.backend_pdf import PdfPages

with PdfPages("Churn_Analysis_Report.pdf") as pdf:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=feat_importance[:10], y=feat_importance.index[:10])
    plt.title("Top 10 Feature Importances")
    plt.tight_layout()
    pdf.savefig()
    plt.close()

    # Add more charts similarly and save
    print("ðŸ“„ PDF Report saved as 'Churn_Analysis_Report.pdf'")
```
